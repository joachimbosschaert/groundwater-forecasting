## Create LSTM models to forecast depth to ground water

## How does LSTM work?
An LSTM is a model that aims to extract information from any type of sequential data. When a time series is passed through an LSTM, the LSTM aims to learn both long term and short term dependencies.

Similarly to a layer in a neural network having nodes, an LSTM layer is made of LSTM cells. The sequential data is passed through each of these LSTM cells and each cell outputs a single value for that time series. In this sense, you can think of an LSTM cell as a filter, similar to a trainable kernel being convolved over an image in a CNN. The LSTM layer outputs a vector of length equal to the number of LSTM cells in that layer.

How does the LSTM cell work?
The LSTM cell is composed of a Cell state, a hidden state. There are a number of gates that aim to update the cell state and hidden state.

<img src="../Images/LSTM_cell_pic.png">
Image taken from "https://colah.github.io/posts/2015-08-Understanding-LSTMs/"


The first gate to discuss is the forget gate. The forget gate decides how much of the cell state to forget. The previous hidden state ($h_{t-1}$) and the current input $({x_t})$ are combined (can be concatenated) and input in a sigmoid function in the following way:

$$f_t = \sigma (W_f \cdot[h_{t-1},x_t] + b_f)$$

The sigmoid function ensures that the output is between zero and one. The output of the forget gate is then multiplied by the previous cell state ($c_{t-1}$) to determine ho much of it to remember. This is represented by the left-most gate in the diagram.

Next we decide how to update the cell state. We use two operations to update cells state. First an input gate, which determines how much of the input to use for the update. This is described by the following equation.

$$i_t = \sigma (W_i \cdot[h_{t-1},x_t] + b_i)$$

This is the same operation as the forget gate, and essentially decides how much of the input and previous hidden state to use for the current operation, but with a different set of weights. This is combined with a tanh layer, that determines how the input affects the cell state, represented as follows:

$$\tilde{C}_t = \tanh (W_C \cdot[h_{t-1},x_t] + b_C)$$

$\tilde{C}_t$ determines how to update the cell state, while $i_t$ determines how much to update the cell state.

The full update of the cell state is represented as follows, and is represented by the center part of the LSTM
cell image.

$$ C_t = f_t*C_{t-1} + i_t*\tilde{C}_t$$

The final part of the computation is updating the hidden state. In this operation the cell state, the previous hidden state, and the input are all used to calculate a new hidden state as follows. This is known as the output gate.

$$ o_t = \sigma (W_o \cdot[h_{t-1},x_t] + b_o)$$
$$ h_t = o_t * \tanh(C_t)$$


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error, mean_squared_error
import math


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, TimeDistributed, BatchNormalization, Dense, Dropout
from tensorflow.random import set_seed
from tensorflow.keras.regularizers import L1L2

import warnings
warnings.filterwarnings('ignore')
```

## Load preprocessed data

```python
df = pd.read_csv("../Processed_data/processed_data.csv", index_col="Unnamed: 0")
df.head()
```

## Use minmax scaler on Depth to groundwater

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler((-1,1))
df["Depth_to_groundwater_scaled"] = scaler.fit_transform(df[["Depth_to_groundwater"]])
```

```python
df.Depth_to_groundwater_scaled
```

## Parameters for the model

```python
## set seed for reproducability
set_seed(5678)

### how many steps for future forecast
len_forecast = 10
look_back = 50

## num folds
k = 10

### num nodes per layer
LSTM_layer1 = 64
LSTM_layer2 = 32
LSTM_layer3 = 32
Dense_layer1 = 32

Dropout_1 = 0.2
Dropout_2 = 0.2
```

## Create training and testing set

```python
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits = k)
```

## Process time series as input for LSTM
For LSTM, similar to any other supervised learning model we need labeled input data.
In this case, one training example is a subset of the time series, with the label being the following value.
For example, if we set the input length of our model to be 50 values, then an X_train would be the first 50 points of the time series, with y_train being the 51st value of the time series

```python
def split_time_series(sequence, num_steps):
    
    X = np.zeros(shape = (len(sequence) - num_steps - 1, num_steps))
    y = np.zeros(shape = len(sequence) - num_steps - 1)
    for i in range(len(sequence) - num_steps - 1):
        X[i] = sequence[i:i+num_steps]
        y[i] = sequence[i+num_steps]
        
    return X, y
```

## Create training set
Create training set to train the model
Create holdout set to compare performance with ARIMA model

```python
start_index = -len_forecast

X = None
y = None

for fold, (train_index, holdout_index) in enumerate(tscv.split(df.Depth_to_groundwater_scaled)):
    train_sequence = train_index[start_index + len_forecast:]
    start_index = train_index[-1]
    
    split_sequence = df.Depth_to_groundwater_scaled[train_index]
    
    seqX, label = split_time_series(split_sequence, look_back)
    
    

    try:
        X = np.concat([X,seqX])
        y = np.concat([y,label])
        
    except:
        X = seqX
        y = label
```

## Split samples into training and testing
- Xtrain & Xtest should have shape (num samples, num features, look_back)

- In our case we have 1 feature and look_back has been set to 50

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.1, random_state=1234)

### Reshape for LSTM
Xtrain = np.reshape(Xtrain, (Xtrain.shape[0], 1, Xtrain.shape[1]))
Xtest = np.reshape(Xtest, (Xtest.shape[0], 1, Xtest.shape[1]))
```

## Add callback  to stop early

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop_callback = EarlyStopping(
    monitor = 'val_loss',
    patience = 50,
    restore_best_weights=True)
```

## A note on overfitting
In previous iterations, the model had been overfitting, Various actions were taken to prevent this from happening
- Adding dropout layers. When dropout was added, both the training and validation loss increased a lot, even for small amounts of dropout.
- Regularizing the weights using L1 and L2 regularization. This performed quite well
- Time distributed batch normalization between LSTM layers
    - This slowed training, and model did not converged to a higher loss
- Tried Batchnormalization on input data, 
    - this significantly slowed training, and model converged to higher loss
- Add callback evaluated on validation loss, find model with the lowest validation loss

## Create the model

```python
#### initialize the regularizer
reg1 = L1L2(l1 = 4e-5, l2 = 4e-5)

#### the model
model = Sequential()

#model.add(BatchNormalization(input_shape=(1, look_back)))         
model.add(LSTM(LSTM_layer1, activation='relu', return_sequences=True, input_shape=(1, look_back), kernel_regularizer=reg1))
#model.add(TimeDistributed(BatchNormalization()))
#model.add(Dropout(0.05))

model.add(LSTM(LSTM_layer2, activation='relu',return_sequences=True))
model.add(LSTM(LSTM_layer3, activation='relu'))

model.add(Dense(Dense_layer1, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

## Fit the model

```python
history = model.fit(Xtrain, ytrain,
                  validation_data=(Xtest, ytest),
                  epochs=500, 
                  callbacks=early_stop_callback,
                  verbose=2)
```

## Plot trainnig & validation loss functions

```python
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = np.arange(1, len(train_loss) + 1)

skipstep = 5

plt.figure(figsize=(10,8))
plt.plot(epochs[skipstep:], train_loss[skipstep:], label = "training loss", linewidth = 2, alpha = 0.9)
plt.plot(epochs[skipstep:], val_loss[skipstep:], label = "validation loss", linewidth = 2, alpha = 0.9)

plt.legend()
```

## Predict next value

```python
#fig, ax = plt.subplots(5,2, figsize = (16,12))

MSE_scores = []
MAE_scores = []

fig, ax = plt.subplots(5,2, figsize = (16,12))

for fold, (train_index, holdout_index) in enumerate(tscv.split(df.Depth_to_groundwater_scaled)):
    
    ax2 = int(fold/5)
    ax1 = int(fold%5)
#     print(ax1, ax2)
    
    #### create holdout sets using split_time_series
    holdout_index = np.concatenate([np.arange(holdout_index[0]-look_back, holdout_index[0]), holdout_index])
    train, holdout = df.Depth_to_groundwater_scaled[train_index], df.Depth_to_groundwater_scaled[holdout_index]
    
    X_holdout, y_holdout = split_time_series(np.array(holdout), look_back)
    X_holdout = X_holdout.reshape(X_holdout.shape[0], 1, look_back)
        
    #### generate models predictions
    predictions = model.predict(X_holdout)
    
    ##scale back to normal
    predictions_inv = scaler.inverse_transform(predictions)

    ### get mse and mae
    mse = mean_squared_error(y_holdout, predictions_inv)
    mae = mean_absolute_error(y_holdout, predictions_inv)

    MSE_scores.append(mse)
    MAE_scores.append(mae)
    
    
    #### plot test and predicted
    ax[ax1, ax2].plot(holdout_index[:], df.Depth_to_groundwater[holdout_index[:]], color="black", label=f"Actual Fold {fold}")  # Test actuals
    ax[ax1, ax2].plot(holdout_index[look_back+1:], predictions_inv[:], color="red", linestyle="dashed", label=f"Predicted Fold {fold}")  # Predictions

    # Confidence interval shading
    #ax[ax1, ax2].fill_between(test_index[:10], upper_conf_trans, lower_conf_trans, color='red', alpha=0.2)
    ax[ax1, ax2].plot(train_index[-200:], df.Depth_to_groundwater[train_index[-200:]], color='blue', label ="original time series")
    ax[ax1, ax2].legend(loc="upper left")
    
fig.suptitle("Next value Predictions vs actuals for all folds", fontsize=20)
plt.tight_layout()
```

## Forecast on test set

```python
def forecast_LSTM(model, last_sequence, look_back, len_forecast, verbose=1):
    
    predictions = []
    LSTM_input = last_sequence.copy()
    
    #print(LSTM_input)
    LSTM_input = np.reshape(LSTM_input, (1, 1, look_back))
    
    #print(LSTM_input)
    for i in range(len_forecast):
        pred_val = model.predict(LSTM_input, verbose = verbose)
        predictions.append(pred_val[0])
                
        ##update the LSTM input sequence
        ##using roll, every value is shifted back by one, and the first value becomes the last value
        ## replace the last value with the predicted value
        LSTM_input = np.roll(LSTM_input, shift=-1)
        LSTM_input[0, 0,-1] = pred_val[0,0]
        
    return predictions
```

```python
#fig, ax = plt.subplots(5,2, figsize = (16,12))
MSE_scores = []
MAE_scores = []

fig, ax = plt.subplots(5,2, figsize = (16,12))

for fold, (train_index, holdout_index) in enumerate(tscv.split(df.Depth_to_groundwater_scaled)):
    
    ax2 = int(fold/5)
    ax1 = int(fold%5)
#     print(ax1, ax2)
    
    #### create holdout sets using split_time_series 
    train, holdout = df.Depth_to_groundwater_scaled[train_index], df.Depth_to_groundwater_scaled[holdout_index]
    
    holdout = holdout[:len_forecast]
    
        
    #### generate models predictions
    last_sequence = train[-look_back:]
    predictions = forecast_LSTM(model, last_sequence, look_back, len_forecast, verbose=0)
    
    ##scale back to normal
    predictions_inv = scaler.inverse_transform(predictions)

    ### get mse and mae
    mse = mean_squared_error(holdout, predictions_inv)
    mae = mean_absolute_error(holdout, predictions_inv)

    MSE_scores.append(mse)
    MAE_scores.append(mae)
    
    
    #### plot test and predicted
    ax[ax1, ax2].plot(holdout_index[:len_forecast], df.Depth_to_groundwater[holdout_index[:len_forecast]], color="black", label=f"Actual Fold {fold}")  # Test actuals
    ax[ax1, ax2].plot(holdout_index[:len_forecast], predictions_inv, color="red", linestyle="dashed", label=f"Predicted Fold {fold}")  # Predictions

    # Confidence interval shading
    #ax[ax1, ax2].fill_between(test_index[:10], upper_conf_trans, lower_conf_trans, color='red', alpha=0.2)
    ax[ax1, ax2].plot(train_index[-200:], df.Depth_to_groundwater[train_index[-200:]], color='blue', label ="original time series", zorder = 1)
    ax[ax1, ax2].legend(loc="upper left")
    
fig.suptitle("Predictions vs actuals for all folds", fontsize=20)
plt.tight_layout()
```

## Parameter tuning
- using grid search find optimal hyper parameter
- This will be evaluated on the test set, not the holdout set

- We will tune Units, L1, and L2,

```python
def create_model(units, L1, L2):
    #### initialize the regularizer
    reg1 = L1L2(l1 = L1, l2 = L2)

    #### the model
    model = Sequential()

    #model.add(BatchNormalization(input_shape=(1, look_back)))         
    model.add(LSTM(units, activation='relu', return_sequences=True, input_shape=(1, look_back), kernel_regularizer=reg1))


    model.add(LSTM(int(units*0.5), activation='relu',return_sequences=True))
    model.add(LSTM(int(units*0.3), activation='relu'))

    model.add(Dense(int(units*0.3), activation = 'relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mse')
    
    return model
```

## Perform grid search
- Tried to use scikeras wrapper Kerasclassifier with sklearn.model_selection.GridSearchCV to perform the grid search, but this only worked for a specific set of parameters, (not *L1* and *L2*)
- Manually perform grid search by looping through all combinations
- Use Kfold cross validation to evaluate model performance on the training set

```python
### define dictionary of search combinations
from itertools import product

param_grid = {
    'units' : [32, 64, 96], 
    'L1' : [2e-5, 3e-5, 4e-5],
    'L2' : [2e-5, 3e-5, 4e-5],
}

param_names = param_grid.keys()
param_vals = param_grid.values()

combinations = [dict(zip(param_names, combo)) for combo in product(*param_vals)]
```

```python
### define Kfolds
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
model_val_loss = {}
best_model = None
best_loss = np.float16('inf')

for combo in combinations[:]:

    loss_vals = []
    for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
        
        #### define training and validation data for each fold
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        
        ### reshape for LSTM input
        X_train = X_train.reshape(X_train.shape[0], 1, look_back)
        X_val = X_val.reshape(X_val.shape[0], 1, look_back)
        
        ### initialize the callback
        early_stop_callback = EarlyStopping(
                                    monitor = 'val_loss',
                                    patience = 50,
                                    restore_best_weights=True)
        
        ## define model, redefine for every fold so the weights are reset
        ## model parameters defined by combo
        model = create_model(combo['units'],combo['L1'],combo['L2'])

        
        ## fit model to data for given fold
        history = model.fit(X_train, y_train,
                              validation_data=(X_val, y_val),
                              epochs=500, 
                              callbacks=early_stop_callback,
                              verbose=0)
        
        ### use validation loss as performance metric
        loss_vals.append(min(history.history["val_loss"]))
    
    ### get the average validation loss for a parameter selection
    mean_loss = np.mean(loss_vals) 
    model_val_loss[str(combo)] = mean_loss
    
    print(combo, np.mean(loss_vals))
    
    ## if its the best model, save it
    if mean_loss < best_loss:
        print("new best model")
        best_loss = mean_loss
        best_model = model
```

## Train model with best parameters
- Use train and test set, not cross validation

```python
### best params
units = 32
L1 = 2e-5
L2 = 2e-5

early_stop_callback = EarlyStopping(
                            monitor = 'val_loss',
                            patience = 50,
                            restore_best_weights=True)

model = create_model(units, L1, L2)

## fit model to data for given fold
history = model.fit(X_train, y_train,
                      validation_data=(X_val, y_val),
                      epochs=500, 
                      callbacks=early_stop_callback,
                      verbose=1)

model.save('./Models/LSTM_1feature.keras')
```
